```python
#Import libraries

import numpy as np
from math import pi
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
# Define variables 
x = np.arange(0, 2*pi, 0.1);
y = np.sin(x)

plt.plot(x, y)

# Add more fancy features
plt.title('First Graph')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid()
```
